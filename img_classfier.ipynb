```python
import os
import cv2
import random
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D
from tensorflow.keras.models import Model
from sklearn.metrics import confusion_matrix, classification_report

# ============================
# 1. Setup & Configuration
# ============================
# Update this path to match your local directory structure
# This assumes your mini-dataset is located here
DATASET_PATH = r'C:\Users\HARI GOVIND A\Downloads\dc_dataset'
TRAIN_DIR = os.path.join(DATASET_PATH, 'train')
VAL_DIR = os.path.join(DATASET_PATH, 'validation')
TEST_DIR = os.path.join(DATASET_PATH, 'test')

# Model configuration
TARGET_SIZE = (224, 224)
BATCH_SIZE = 32
NUM_EPOCHS = 10
NUM_CLASSES = 1 # Binary classification for Cats vs Dogs

# ============================
# 2. Dataset Exploration (Phase 1)
# ============================
def explore_dataset():
    """
    Explores the dataset structure and displays a random image.
    """
    print("📊 Exploring Dataset Structure...")
    categories = os.listdir(TRAIN_DIR)
    print(f"Categories found: {categories}")

    for category in categories:
        path = os.path.join(TRAIN_DIR, category)
        num_images = len(os.listdir(path))
        print(f"Number of images in '{category}': {num_images}")

    # Display a random image
    random_category = random.choice(categories)
    random_path = os.path.join(TRAIN_DIR, random_category)
    random_image_name = random.choice(os.listdir(random_path))
    image_path = os.path.join(random_path, random_image_name)
    
    img = cv2.imread(image_path)
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    
    print("\n--- Displaying a Random Image ---")
    print(f"Image Path: {image_path}")
    print(f"Image Shape: {img_rgb.shape}")
    
    plt.imshow(img_rgb)
    plt.title(f"Random Image from '{random_category}' category")
    plt.axis('off')
    plt.show()

# ============================
# 3. Data Preprocessing & Augmentation (Phase 2)
# ============================
def create_data_generators():
    """
    Sets up ImageDataGenerators for training, validation, and testing
    with data augmentation for the training set.
    """
    print("🔄 Setting up Data Generators...")
    # Generator for training data with augmentation and normalization
    train_datagen = ImageDataGenerator(
        rescale=1./255, 
        rotation_range=20,
        width_shift_range=0.2,
        height_shift_range=0.2,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True,
        fill_mode='nearest'
    )

    # Generators for validation and test data (only normalization)
    val_test_datagen = ImageDataGenerator(rescale=1./255)

    train_generator = train_datagen.flow_from_directory(
        TRAIN_DIR,
        target_size=TARGET_SIZE,
        batch_size=BATCH_SIZE,
        class_mode='binary'
    )
    
    validation_generator = val_test_datagen.flow_from_directory(
        VAL_DIR,
        target_size=TARGET_SIZE,
        batch_size=BATCH_SIZE,
        class_mode='binary'
    )

    test_generator = val_test_datagen.flow_from_directory(
        TEST_DIR,
        target_size=TARGET_SIZE,
        batch_size=1, # Use a batch size of 1 for evaluation
        class_mode='binary',
        shuffle=False
    )
    
    return train_generator, validation_generator, test_generator

# ============================
# 4. Model Building (Phase 3)
# ============================
def build_model():
    """
    Builds a CNN model using the Transfer Learning approach with MobileNetV2.
    """
    print("🧠 Building the CNN Model...")
    # Load the MobileNetV2 base model, pre-trained on ImageNet
    base_model = MobileNetV2(input_shape=(224, 224, 3), include_top=False, weights='imagenet')
    
    # Freeze the layers of the base model so they don't get retrained
    base_model.trainable = False
    
    # Add new classification layers on top of the base model
    x = base_model.output
    x = GlobalAveragePooling2D()(x)
    x = Dense(128, activation='relu')(x)
    predictions = Dense(NUM_CLASSES, activation='sigmoid')(x)
    
    model = Model(inputs=base_model.input, outputs=predictions)
    
    # Compile the model
    model.compile(optimizer='adam',
                  loss='binary_crossentropy',
                  metrics=['accuracy'])
                  
    model.summary()
    return model

# ============================
# 5. Training & Evaluation (Phase 4 & 5)
# ============================
def train_and_evaluate_model(model, train_gen, val_gen, test_gen):
    """
    Trains the model and evaluates its performance.
    """
    print("\n🚀 Starting Model Training...")
    history = model.fit(
        train_gen,
        epochs=NUM_EPOCHS,
        validation_data=val_gen,
        steps_per_epoch=train_gen.samples // BATCH_SIZE,
        validation_steps=val_gen.samples // BATCH_SIZE
    )
    
    print("\n✅ Training Complete! Evaluating Model on Test Data...")
    
    # Evaluate on the test set
    test_loss, test_acc = model.evaluate(test_gen, steps=test_gen.samples)
    print(f"\nFinal Test Accuracy: {test_acc:.4f}")
    
    # Get a detailed classification report and confusion matrix
    y_pred_probs = model.predict(test_gen)
    y_pred = (y_pred_probs > 0.5).astype(int)
    y_true = test_gen.classes

    print("\n--- Classification Report ---")
    print(classification_report(y_true, y_pred, target_names=['cats', 'dogs']))

    print("\n--- Confusion Matrix ---")
    cm = confusion_matrix(y_true, y_pred)
    print(cm)
    
    # Plotting the training history
    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'], label='Training Accuracy')
    plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
    plt.title('Training and Validation Accuracy')
    plt.legend()
    
    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'], label='Training Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.title('Training and Validation Loss')
    plt.legend()
    
    plt.show()
    
    return model
    
# ============================
# 6. Prediction Function (Phase 6)
# ============================
def predict_image_sentiment(model, image_path):
    """
    Loads an image from a given path, preprocesses it, and predicts its class.
    """
    print("\n🎤 Making a prediction...")
    try:
        img = Image.open(image_path).convert('RGB')
        img = img.resize(TARGET_SIZE)
        img_array = np.array(img) / 255.0  # Normalization
        img_array = np.expand_dims(img_array, axis=0) # Add batch dimension
        
        prediction = model.predict(img_array)[0][0]
        
        if prediction > 0.5:
            sentiment = 'Dog'
            confidence = prediction
        else:
            sentiment = 'Cat'
            confidence = 1 - prediction

        print(f"Prediction: {sentiment.upper()} (Confidence: {confidence:.2%})")
        plt.imshow(img)
        plt.title(f"Prediction: {sentiment.upper()}")
        plt.axis('off')
        plt.show()

    except FileNotFoundError:
        print(f"Error: The file at {image_path} was not found.")
    except Exception as e:
        print(f"An error occurred: {e}")

# ============================
# Main Execution Block
# ============================
if __name__ == "__main__":
    # Ensure the directories exist before running
    if not all(os.path.isdir(d) for d in [TRAIN_DIR, VAL_DIR, TEST_DIR]):
        print("Error: Dataset directories not found. Please update DATASET_PATH.")
    else:
        # Phase 1: Exploration
        explore_dataset()

        # Phase 2: Data Preparation
        train_generator, validation_generator, test_generator = create_data_generators()

        # Phase 3: Model Building
        model = build_model()
        
        # Phase 4 & 5: Training & Evaluation
        trained_model = train_and_evaluate_model(model, train_generator, validation_generator, test_generator)

        # Phase 6: Interactive Prediction (for a single image)
        # Replace 'path/to/your/image.jpg' with a real image path
        # predict_image_sentiment(trained_model, 'path/to/your/image.jpg')
```
